```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from collections import OrderedDict
from torch.utils.data import Dataset, DataLoader

from typing import List, Tuple, Dict, Optional 

from flwr.common import ndarrays_to_parameters, Context
from flwr.common import Metrics, NDArrays, Scalar
from flwr.client import Client, ClientApp, NumPyClient
from flwr.server import ServerApp, ServerConfig, ServerAppComponents
from flwr.server.strategy import FedAvg
from flwr.simulation import run_simulation

from logging import ERROR 

import sys 
sys.path.append("../") 

from src.paths import DATA_DIR, CLIENTS_DATA_DIR
```

```python
from sklearn.metrics import (
    f1_score, precision_score, recall_score, roc_auc_score, 
    confusion_matrix, classification_report, roc_curve
)

final_result = []

def final_test_evaluation(model, test_set):
    """Evaluate model on test data and plot the ROC curve."""
    model.eval()
    test_loader = DataLoader(test_set, batch_size=32, shuffle=False)

    all_preds = []
    all_probs = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            probs = torch.softmax(outputs, dim=1)[:, 1]  # Get probability of class 1
            _, predicted = torch.max(outputs, 1)

            all_preds.extend(predicted.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Compute metrics
    f1 = f1_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds)
    recall = recall_score(all_labels, all_preds)
    roc_auc = roc_auc_score(all_labels, all_probs)
    cm = confusion_matrix(all_labels, all_preds)

    # Print Metrics
    print("\nðŸ“Œ **Final Test Set Metrics** ðŸ“Œ")
    print("F1 Score:", f1)
    print("Precision:", precision)
    print("Recall:", recall)
    print("ROC-AUC Score:", roc_auc)
    print("\nConfusion Matrix:\n", cm)
    print("\nClassification Report:\n", classification_report(all_labels, all_preds))

    # Plot ROC Curve
    fpr, tpr, _ = roc_curve(all_labels, all_probs)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f"ROC curve (area = {roc_auc:.2f})")
    plt.plot([0, 1], [0, 1], linestyle="--", color="gray")  # Random classifier line
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("Receiver Operating Characteristic (ROC) Curve")
    plt.legend(loc="lower right")
    plt.show()

    return f1, precision, recall, roc_auc, cm
```

```python
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
# ===========================
# ðŸ“Œ Step 1: Load Dataset
# ===========================

class FraudDataset(Dataset):
    def __init__(self, x_file, y_file):
        self.x_data = pd.read_csv(x_file).values  # Load features
        self.y_data = pd.read_csv(y_file).values.squeeze()  # Load labels
        
        # Convert to tensors
        self.x_data = torch.tensor(self.x_data, dtype=torch.float32)
        self.y_data = torch.tensor(self.y_data, dtype=torch.long)  # Ensure long dtype for classification

    def __len__(self):
        return len(self.x_data)

    def __getitem__(self, idx):
        return self.x_data[idx], self.y_data[idx]


# Function to load client datasets
def load_client_data(client_id):
    data_dir = CLIENTS_DATA_DIR
    x_train = data_dir/f"x_train_{client_id}.csv"
    y_train = data_dir/f"y_train_{client_id}.csv"
    x_val = data_dir/f"X_val_{client_id}.csv"
    y_val = data_dir/f"y_val_{client_id}.csv"

    train_set = FraudDataset(x_train, y_train)
    val_set = FraudDataset(x_val, y_val)
    
    return train_set, val_set


# Load global test set
def load_test_data():
    data_dir = CLIENTS_DATA_DIR
    return FraudDataset(data_dir/"X_test.csv", data_dir/"y_test.csv")
```

```python
num_clients = 5 #link this to read from config file later
train_sets = []
val_sets = []

for i in range(num_clients):
    train_set, val_set = load_client_data(i)
    train_sets.append(train_set)
    val_sets.append(val_set)

test_set = load_test_data()
```

```python
# ===========================
# ðŸ“Œ Step 2: Define Model
# ===========================
class FraudDetectionModel(nn.Module):
    def __init__(self, num_features):
        super(FraudDetectionModel, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(num_features, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 2)  # Output layer (2 classes: fraud or not)
        )
    
    def forward(self, x):
        return self.model(x)
```

```python
# ===========================
# ðŸ“Œ Step 3: Training & Evaluation Functions
# ===========================
def train_model(model, train_set):
    batch_size = 32
    num_epochs = 10

    train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    model.train()
    for epoch in range(num_epochs):
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()


def evaluate_model(model, val_set):
    model.eval()
    val_loader = DataLoader(val_set, batch_size=32, shuffle=False)
    criterion = nn.CrossEntropyLoss()

    correct, total, total_loss = 0, 0, 0.0
    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)
            total_loss += criterion(outputs, labels).item()

    accuracy = correct / total
    avg_loss = total_loss / len(val_loader)
    return avg_loss, accuracy
```

```python
# ===========================
# ðŸ“Œ Step 4: Fix Model Weight Serialization
# ===========================

# Retrieves the parameters from the model
def get_weights(model):
    """Extract model weights and return them as a Flower Parameters object."""
    nd_array_weights = [val.cpu().numpy() for _, val in model.state_dict().items()]
    return nd_array_weights

# Sets the parameters of the model
def set_weights(model, parameters):
    params_dict = zip(model.state_dict().keys(), parameters)
    state_dict = OrderedDict({k: torch.tensor(v) for k, v in params_dict})
    model.load_state_dict(state_dict, strict=True)
```

```python
# ===========================
# ðŸ“Œ Step 5: Define Flower Client
# ===========================
class BankClient(NumPyClient):
    def __init__(self, model, trainset, valset):
        self.model = model
        self.trainset = trainset
        self.valset = valset
        
    # Train the model
    def fit(self, parameters, config):
        set_weights(self.model, parameters)
        train_model(self.model, self.trainset)
        return get_weights(self.model), len(self.trainset), {}
        
    # Test the model
    def evaluate(self, parameters: NDArrays, config: Dict[str, Scalar]):
        set_weights(self.model, parameters)
        loss, accuracy = evaluate_model(self.model, self.valset)
        return loss, len(self.valset), {"accuracy": accuracy}
```

```python
# ===========================
# ðŸ“Œ Step 6: Define the Client Function
# ===========================

def client_fn(context: Context) -> Client:
    """Create a federated learning client."""
    partition_id = int(context.node_config["partition-id"])  # Get the client partition ID
    model = FraudDetectionModel(num_features=51)  # Create model
    client_train = train_sets[partition_id]  # Assign partitioned training dataset
    client_val = val_sets[partition_id]  # Assign partitioned validation dataset
    
    return BankClient(model, client_train, client_val).to_client()

client = ClientApp(client_fn)
```

```python
# ===========================
# ðŸ“Œ Step 7: Define the Server and Run Training
# ===========================

def evaluate(server_round, parameters, config):
    """
    Evaluate the global model on the test set after each round.
    """
    model = FraudDetectionModel(num_features=51)
    set_weights(model, parameters)

    loss, accuracy = evaluate_model(model, test_set)

    # Run detailed evaluation **only after the last round**
    if server_round == 5:  
        f1, precision, recall, roc_auc, cm = final_test_evaluation(model, test_set)

    return loss, {"accuracy": accuracy}
```

```python
def weighted_average(metrics: List[Tuple[int, Metrics]]) -> Metrics:
    # Multiply accuracy of each client by number of examples used
    accuracies = [num_examples * m["accuracy"] for num_examples, m in metrics]
    examples = [num_examples for num_examples, _ in metrics]

    # Aggregate and return custom metric (weighted average)
    return {"accuracy": sum(accuracies) / sum(examples)}
```

```python
model = FraudDetectionModel(num_features=51)
params = ndarrays_to_parameters(get_weights(model))

def server_fn(context: Context):
    strategy = FedAvg(
        fraction_fit=1.0,                             #<--- Sample 100% of available clients for training
        fraction_evaluate=1.0,
        initial_parameters=params,
        evaluate_fn=evaluate, 
        evaluate_metrics_aggregation_fn=weighted_average,  # <-- pass the metric aggregation function
    )
    
    sever_config = ServerConfig(num_rounds=5) #<--- no. of federated rounds
    return ServerAppComponents(strategy = strategy, config = sever_config)


server = ServerApp(server_fn = server_fn)
```

-----

```python
backend_setup = {
    # Filter logging coming from the Simulation Engine so it's more readable in notebooks
    "init_args": {"logging_level": ERROR, "log_to_driver": False},
    # By default, each client will be allocated 1x CPU and 0x GPUs
    "client_resources": {"num_cpus": 1, "num_gpus": 0.0}
    }

# When running on GPU, assign an entire GPU for each client
if DEVICE == "cuda": 
    backend_setup["client_resources"] = {"num_cpus": 1, "num_gpus": 1.0}
```

```python
# Initiate the simulation passing the server and client apps
# Specify the number of super nodes that will be selected on every round

run_simulation(
    server_app=server,
    client_app=client,
    num_supernodes=5, # number of clients link this to config file later
    backend_config=backend_setup,
)
```
